```python
# dropna drops missing values (na as "not available"

df1 = df.dropna(axis=0)
df1.shape
```

```python
# df1 is the df without null values

df1.isnull().sum()
```

```python
df1.shape
```

```python
s = (df.dtypes == 'object')
object_cols = list(s[s].index)

print("Categorical variables:")
print(object_cols)
```

```python
y = df1.Price
```

```python
features = ['Rooms', 'Bedroom2', 'Bathroom', 'Landsize', 'YearBuilt', 'Lattitude', 'Longtitude']
```

```python
X = df1[features]
X.describe()
```

```python
X.dtypes
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
# extreme gradient method

from xgboost import XGBRegressor

xgb_model = XGBRegressor(n_estimators=500)
xgb_model.fit(X_train, y_train)
```

```python
print(X.head())
print(xgb_model.predict(X.head()))
```

```python
#error = actual - predicted

#MAE - mean adsolute value

from sklearn.metrics import mean_absolute_error

predicted_home_prices = xgb_model.predict(X)
mean_absolute_error(y, predicted_home_prices)
```
